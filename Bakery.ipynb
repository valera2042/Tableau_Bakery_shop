```python
dateTime=pd.to_datetime(bakeryDF['DateTime'])
bakeryDF['Day']=dateTime.dt.day_name()
bakeryDF['Month']=dateTime.dt.month_name()
bakeryDF['Year']=dateTime.dt.year
bakeryDF.head(5)
```

```python
bakeryDF['hour']=dateTime.dt.hour
```

```python
bakeryDF
```

```python
bakeryDF['part_of_the_day'] = 'morning'
```

```python
def get_the_day(x):

    

    if 0 <= x < 2:
        return '12am-2am'
    if 2 <= x < 6:
        return '2am-6am'
    if 6 <= x < 8:
        return '6am-8am'
    if 8 <= x < 11:
        return '8am-11am'
    if 11 <= x < 12:
        return '11am-12pm'
    if 12 <= x < 15:
        return '12pm-3pm'
    if 15 <= x < 17:
        return '3pm-5pm'
    if 17 <= x < 18:
        return '5pm-7pm'
    if 18 <= x < 21:
        return '7pm-9pm'
    if 21 <= x < 23:
        return '9pm-11pm'
    if 23 <= x < 0:
        return '11pm-12am'
```

```python
bakeryDF['hour'] = bakeryDF['hour'].astype(int)
```

```python
bakeryDF['part_of_the_day'] = bakeryDF['hour'].apply(lambda x: get_the_day(x))
```
